```python
import pandas as pd
import numpy as np
```

```python
application_data = pd.read_csv('application_data.csv')
previous_data = pd.read_csv('previous_application.csv')
```

```python
application_data.shape
```

```python
def correlated(dataset, threshold):
    corr_columns = set()
    corr_ = dataset.corr()
    for i in range(len(corr_.columns)):
        for j in range(i):
            if abs(corr_.iloc[i,j]) >= threshold:
                corr_columns.add(corr_.columns[i])
                
    return corr_columns
```

```python
app_corr_cols = correlated(application_data, 0.6)
```

```python
application_data = application_data.drop(app_corr_cols, axis=1)
```
